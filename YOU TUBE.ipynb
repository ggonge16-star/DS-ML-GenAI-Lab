# Taxt Clining

```python
df['title']=df['title'].str.upper()
df['tags']=df['tags'].str.upper()
```

```python
#Special Characters
df['title'] =df['title'].str.replace(r'[^a-zA-Z ]', '', regex=True)
df['tags'] =df['tags'].str.replace(r'[^a-zA-Z ]', '', regex=True)
df['channel_title'] = df['channel_title'].str.replace(r'[^a-zA-Z]','',regex=True)
df['time_frame'] = df['time_frame'].str.replace('to','',regex=True)
```

```python
df['tags']=df['tags'].str.upper()
```

```python
#Multiple Spaces
df['tags'] = df['tags'].str.replace(r'\s+', ' ', regex=True)
```

# 10. Visualization (Matplotlib / Seaborn)

### Views distribution histogram.

```python
plt.hist(df["views"], bins=50)
plt.xlabel("Views")
plt.ylabel("Frequency")
plt.title("Views Distribution Histogram")
plt.show()
```

### Top 10 channels bar chart.


```python
top_channels = (df["channel_title"].value_counts() .head(10))
top_channels.plot(kind="bar", figsize=(10,5))
plt.title("Top 10 Channels with Most Trending Videos")
plt.xlabel("Channel")
plt.ylabel("Number of Trending Videos")
plt.show()
```

### Country-wise videos bar chart.

```python
country_count=df["publish_country"].value_counts()
country_count.plot(kind="bar", figsize=(10,5))
plt.title("Number of Trending Videos by Country")
plt.xlabel("Channel")
plt.ylabel("Number of Trending Videos")
plt.show()
```

### Likes vs Dislikes scatter plot.

```python
# Using matplot
plt.figure(figsize=(8,6))
plt.scatter(df["likes"], df["dislikes"], alpha=0.5)
plt.xlabel("Likes")
plt.ylabel("Dislikes")
plt.title("Likes vs Dislikes Scatter Plot")
plt.show()
```

```python
# Using seaborn
sns.scatterplot(x="likes", y="dislikes", data=df,hue="category_id")
plt.title("Likes vs Dislikes Scatter Plot")
plt.show()
```
